```python
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline

plt.rcParams['font.family'] = 'serif'
plt.rcParams['mathtext.fontset'] = 'dejavuserif'
```

```python
fff = np.linspace(50, 230)
sbw = 23175*(fff/1000)**(22/5)
st = 820.92*(fff/1000)**(6/5) / 60

fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(fff, sbw)
ax[1].plot(fff, st)
ax[1].set_xlabel('frequency [MHz]')
ax[0].set_ylabel('diffractive scintillation\nbandwidth [kHz]')
ax[1].set_ylabel('diffractive scintillation\ntimescale [min]')
ax[0].set_title('PSR J0630$-$2834')
plt.subplots_adjust(hspace=0.1)
plt.show()
```

```python
mb = np.sqrt(0.773*(sbw*1000/(fff*1000000))**(5/6))
plt.plot(fff, mb**2)
#plt.ylim(0, 0.008)
plt.xlabel('frequency [MHz]')
plt.ylabel('Born variance $m_b^2$')

plt.show()
```

```python
plt.plot(fff, (1-0.38)*1/np.sqrt(sbw*1000/(fff*1000000)))
plt.xlabel('frequency [MHz]')
plt.ylabel('Fresnel scale [dimensionless]')
plt.show()
```
